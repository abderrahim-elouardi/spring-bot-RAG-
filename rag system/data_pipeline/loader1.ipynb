```python
import requests
from bs4 import BeautifulSoup
```

```python
urls = [
    ("https://spring.io/guides/gs/accessing-data-jpa", "Accessing Data with JPA"),
    ("https://spring.io/guides/gs/rest-service/", "Building a RESTful Web Service"),
    ("https://spring.io/guides/gs/spring-boot/", "Building an Application with Spring Boot"),
    ("https://spring.io/guides/gs/consuming-rest/", "Consuming a RESTful Web Service"),
    ("https://spring.io/guides/gs/messaging-stomp-websocket/", "Messaging with STOMP over WebSocket"),
    ("https://spring.io/guides/gs/spring-security/", "Securing a Web Application with Spring Security"),
    ("https://spring.io/guides/gs/spring-boot-docker/", "Building a Docker Image with Spring Boot"),
    ("https://spring.io/guides/gs/spring-boot-kubernetes/", "Deploying a Spring Boot Application to Kubernetes"),
    ("https://spring.io/guides/gs/spring-boot-graphql/", "Building a GraphQL API with Spring Boot"),
    ("https://spring.io/guides/gs/spring-boot-testing/", "Testing a Spring Boot Application"),
    ("https://spring.io/guides/gs/spring-boot-actuator/", "Monitoring and Managing Your Application with Spring Boot Actuator")
]

data = []
```

```python
for section in urls:
    request = requests.get(section[0])
    if request.status_code == 200:
        soup = BeautifulSoup(request.content, "html.parser")
        data.append([section[1], soup.get_text(separator=" " , strip=True)])
        print(f"{section[1]}: {len(soup.get_text())} characters")
```

```python
import numpy as np
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
import re
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords


# nltk.download('punkt')       # pour la tokenisation
nltk.download('punkt_tab')       # pour la tokenisation
nltk.download('stopwords')   # pour les mots vides
nltk.download('wordnet')     # pour lemmatisation
nltk.download('omw-1.4')     # lexique pour WordNet
```

```python
stop_words = set(stopwords.words('english'))
def clean_text(text):
    text = text.lower()  # minuscule
    text = re.sub(r"http\S+", "", text)  # supprimer URLs
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)  # enlever ponctuation
    text = re.sub(r"\s+", " ", text).strip()  # enlever espaces en trop
    return text

def prepare_text(text):
    text = clean_text(text)

    #tokenisation
    tokens = word_tokenize(clean_text(text))
    
    #supprimer les stop words
    tokens = [word for word in tokens if word not in stop_words]

    #lammatization 
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    
    #reconstuire le text
    return " ".join(tokens)

```

```python
for i in range(len(data)):
    print(f"Title: {data[i][0]}, Text length: {len(data[i][1])} characters")
    data[i][1] = prepare_text(data[i][1])
```

```python
from pathlib import Path

for i in range(len(data)):
    print(f"Title: {data[i][0]}, Cleaned Text length: {len(data[i][1])} characters")
    title , text = data[i]

    # 1. Définir le dossier et le nom du fichier
    dossier = Path("../data/") # Ou "donnees/resultats" pour un chemin relatif
    nom_fichier = f"{title.replace(' ', '_')}.txt"  # Remplace les espaces par des underscores pour le nom du fichier

    # 2. Créer le dossier s'il n'existe pas (parents=True crée aussi les dossiers intermédiaires)
    dossier.mkdir(parents=True, exist_ok=True)

    # 3. Combiner le dossier et le fichier
    chemin_complet = dossier / nom_fichier

    # 4. Écrire le contenu
    with open(chemin_complet, "w", encoding="utf-8") as f:
        f.write(text)

    print(f"Fichier enregistré ici  : {chemin_complet}")
```
